```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("dark_background")
```

```python
import pandas as pd
df = pd.read_csv("zomato.csv")
df.head()
```

```python
# shape is to see how many rows and columns are there:
df.shape
```

```python
# we see different columns are there in dataset
df.columns
```

```python
df = df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis=1)
df.head()
```

```python
# info() is used provied informantion about my dataset
df.info()
```

# Dropping Duplicates 

```python
#removing duplicates values
df.drop_duplicates(inplace = True)
df.shape
```

# Cleaning rate Column

```python
# arrange the rate column in unique
df['rate'].unique()
```

# Removing "NEW","-"and "/5" from the rate column 


```python
def handlingvalue(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split("/")
        value = value[0]
        return float(value)
df['rate']= df['rate'].apply(handlingvalue)
df['rate'].head()
```

```python
# Checkthe how many null are there in rate column
df['rate'].isnull().sum()
```

# Filling Null values into rate column

```python
# Inplace = true is create a new date frame
df['rate'].fillna(df['rate'].mean(),inplace = True)
df['rate'].isnull().sum()
```

```python
df.info()
```

# Dropping Null Values

```python
df.dropna(inplace = True)
df.head()
```

```python
# Rename is change one name to another name 
df.rename(columns = {'approx_cost(for two people)':"cost2plates",'listed_in(type)':'type'},inplace=True)
df.head()
```

# Now we find the Location

```python
df['location'].unique()
```

```python
df.rename(columns={'listed_in(city)':'locations2'},inplace=True)
df.head()
```

```python
df['locations2'].unique()
```

# locations2 and location are both there lets kep only one

```python
df= df.drop(['locations2'],axis=1)
```

```python
df.head()
```

```python
df['cost2plates'].unique()
```

# Removing "," in the cost2plate column

```python
def handling(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)
df['cost2plates'] = df['cost2plates'].apply(handling)
df['cost2plates'].unique()
```

```python
df.head(10)
```

```python
df['rest_type'].value_counts()
```

```python
restvalue = df['rest_type'].value_counts(ascending = False)
restvalue
```

```python
resttypelessthan1000 = restvalue[restvalue<1000]
resttypelessthan1000
```

```python
def dotted (value):
    if value in resttypelessthan1000:
        return 'others'
    else:
        return value
df['rest_type'] = df['rest_type'].apply(dotted)
df['rest_type'].value_counts()
```

```python
df.head()
```

# Cleaning location column

```python
located = df['location'].value_counts(ascending = False)
located
```

```python
locatedgreaterthan300 = located[located<300]
locatedgreaterthan300
```

```python
def dotted2 (value):
    if value in locatedgreaterthan300:
        return 'others'
    else:
        return value
df['location'] = df['location'].apply(dotted2)
df['location'].value_counts()
```

```python
df.head()
```

# cleaning cuisines Column

```python
cuisines1 = df['cuisines'].value_counts(ascending=False)
cuisines1
```

```python
cuisines2 = cuisines1[cuisines1<100]
cuisines2
```

```python
def ball(value):
    if value in cuisines2:
        return 'others'
    else:
        return value
df['cuisines'] = df['cuisines'].apply(ball)
df['cuisines'].value_counts()
        
```
